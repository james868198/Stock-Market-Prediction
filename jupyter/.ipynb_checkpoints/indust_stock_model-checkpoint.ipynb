```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.utils import normalize
from tensorflow.keras.callbacks import EarlyStopping

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
apple = pd.read_csv('../data/transactions/Tech/AAPL.csv')  
# google = pd.read_csv('../data/transactions/Tech/GOOG.csv')  
# fb = pd.read_csv('../data/transactions/Tech/FB.csv')  
# microsoft = pd.read_csv('../data/transactions/Tech/MSFT.csv')  
# amazon = pd.read_csv('../data/transactions/Tech/AMZN.csv')  
# print(apple)
```

```python

def normalization(arr):
    return (arr- np.mean(arr)) / (np.max(arr) - np.min(arr))
def deDormalization(arr,mean,ma,mi):
    return (arr*(ma-mi)+ mean)
def getStock(stock):
    
    date = pd.to_datetime(stock["Date"])
    
    year = np.array(date.dt.year)/3000
    year = year.reshape(year.shape[0],1)
    month = np.array(date.dt.month)/12
    month = month.reshape(month.shape[0],1)
    day = np.array(date.dt.day)/30
    day = day.reshape(day.shape[0],1)
   
    openData = np.array(stock['Open'])
    openData = normalization(openData)
    openData = openData.reshape(openData.shape[0],1)
    High = np.array(stock['High'])
    High = normalization(High)
    High = High.reshape(High.shape[0],1)
    Low = np.array(stock['Low'])
    Low = normalization(Low)
    Low = Low.reshape(Low.shape[0],1)
    volume = np.array(stock['Volume'])
    volume = normalization(volume)
    volume = volume.reshape(volume.shape[0],1)
    
    output = np.concatenate((year,month,day,openData,High,Low,volume), axis=1)
    CloseData = np.array(stock['Close'])
    
    return output, CloseData

def processData(data,result,input_unit,output_unit):
    X,Y = [],[]
    for i in range(len(data)-input_unit-1):
        X.append(data[i:(i+input_unit)])
        Y.append(result[(i+input_unit):(i+input_unit+output_unit)])
    return np.array(X),np.array(Y)

```

```python
#model parameters
cell_unit = 16
learn_rate = 1e-2
decay_rate = 1e-5
train_epochs = 300

#input parameters
intput = apple
time_num = 3
output_num = 3
precentage = 0.75
```

```python

mean =  np.mean(intput)
ma =  np.max(intput)
mi =  np.min(intput)
data,result = getStock(intput)
data,result = processData(data,result,time_num,output_num)
result.reshape(result.shape[0],1)
samples_num = data.shape[1]
features_num = data.shape[2]

n = len(data)
train_size = int(n*precentage)
data_train,data_test = data[0:train_size],data[train_size:]
result_train,result_test = result[0:train_size],result[train_size:]
print(data_train.shape)
print(result_train.shape)
```
